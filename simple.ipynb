```python
import numpy as np
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
# TARGET_PTM = '(UniMod:37)'
# TARGET_PTM = '(UniMod:7)'
# TARGET_PTM = 'Hypusine'
# TARGET_PTM = 'Deoxyhypusine'
TARGET_PTM = 'acetaldehyde'

PTM_DATA = '/Users/cgu3/Documents/diann/data/PROJ0093_qvalue005/UNIPROT_human_revi_2024_12_19_ProteinAG_acetaldehyde_report-lib.predicted.speclib/PROJ0093_MDAMB231_DMSO_R1_S1-A7_1_11478/PROJ0093_MDAMB231_DMSO_R1_S1-A7_1_11478_report.tsv'
```

```python
metadata = pd.read_excel('data/EI_key_rainbow.xlsx')
```

```python
def extract_ipas_part(filename):
    pattern = r'(IPAS.*?)\.htrms'
    match = re.search(pattern, filename)
    return match.group(1) if match else None

# Apply the function to the 'FileName' column
metadata['PLateNumber'] = metadata['PLateNumber'].apply(extract_ipas_part)
metadata
```

```python
metadata['Cancer Type'].value_counts()
```

```python
citrullination_raw = pd.read_csv(PTM_DATA, sep='\t')
```

```python
citrullination_raw
```

```python
# left join the metadata with the citrullination data
citrullination = citrullination_raw.merge(metadata, left_on='Run', right_on='PLateNumber', how='left')
```

```python
citrullination
```

```python
# peptide-level sum intensity for each run and color by cancer type, only citrullination PTM
```

```python
import re

# check total number of row where the Modified.Sequence contain "(*)" and * means any number of any character
citrullination['has_target_PTM'] = citrullination['Modified.Sequence'].str.contains(TARGET_PTM)
citrullination['num_PTM'] = citrullination['Modified.Sequence'].str.count(TARGET_PTM)
```

```python
citrullination.columns
```

```python
total_intensity = citrullination.groupby(['Run', 'has_target_PTM']).agg(total_intensity=('Precursor.Quantity', 'sum')).reset_index()
total_intensity['total_intensity'] = np.log10(total_intensity['total_intensity'])
total_intensity = total_intensity.merge(metadata, left_on='Run', right_on='PLateNumber')
```

```python
# Drop rows with missing 'Is_Case' values
total_intensity_labeled = total_intensity.dropna(subset=['Is_Case'])

# Aggregate the data by summing the intensities for 'has_target_PTM' False and True
aggregated_intensity = total_intensity_labeled.groupby(['Is_Case', 'Run']).agg(
    total_intensity=('total_intensity', 'sum')
).reset_index()

# Define colors for 'Is_Case'
case_colors = ['blue', 'orange']

# Plot histograms in the same plot with different colors
plt.figure(figsize=(10, 6))
for case_label, case_color in zip(aggregated_intensity['Is_Case'].unique(), case_colors):
    subset = aggregated_intensity[aggregated_intensity['Is_Case'] == case_label]
    subset['total_intensity'].hist(bins=100, alpha=0.5, label=f'Is_Case {case_label}', color=case_color)

plt.xlabel('Total intensity (log10)')
plt.ylabel('Frequency')
plt.title('Distribution of total intensity by Case/Control')
plt.legend()
plt.show()
```

```python
# Drop rows with missing 'Is_Case' values
total_intensity_labeled = total_intensity.dropna(subset=['Is_Case'])

# Define colors for 'Is_Case' and 'has_target_PTM'
case_colors = ['blue', 'orange']
citru_colors = ['green', 'red']

# Plot histograms in the same plot with different colors
plt.figure(figsize=(10, 6))
for case_label, case_color in zip(total_intensity_labeled['Is_Case'].unique(), case_colors):
    for citru_label, citru_color in zip(total_intensity_labeled['has_target_PTM'].unique(), citru_colors):
        subset = total_intensity_labeled[(total_intensity_labeled['Is_Case'] == case_label) &
                                         (total_intensity_labeled['has_target_PTM'] == citru_label)]
        subset['total_intensity'].hist(bins=100, alpha=0.5,
                                       label=f'Is_Case {case_label}, has_target_PTM {citru_label}',
                                       color=case_color if citru_label else citru_color)

plt.xlabel('Total intensity (log10)')
plt.ylabel('Frequency')
plt.title('Distribution of total intensity')
plt.legend()
plt.show()
```

```python
# Drop rows with missing 'Cancer Type' values
total_intensity_labeled = total_intensity[total_intensity['has_target_PTM']].dropna(subset=['Cancer Type'])

# Plot histograms in the same plot with different colors
plt.figure(figsize=(10, 6))
colors = plt.cm.get_cmap('tab10', len(total_intensity_labeled['Cancer Type'].unique()))

for i, (label, color) in enumerate(zip(total_intensity_labeled['Cancer Type'].unique(), colors.colors)):
    subset = total_intensity_labeled[total_intensity_labeled['Cancer Type'] == label]
    subset['total_intensity'].hist(bins=30, alpha=0.5, label=f'Cancer Type {label}', color=color)

plt.xlabel('Total intensity (log10)')
plt.ylabel('Frequency')
plt.title('Distribution of total intensity by Cancer Type')
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt

# Drop rows with missing 'Cancer Type' values
total_intensity_labeled = total_intensity[total_intensity['has_target_PTM']].dropna(subset=['Cancer Type'])

# Filter for specific cancer types and controls
desired_types = ['Gastric', 'Breast', 'Lung', 'LEAP Control', 'MERIT Control']
total_intensity_labeled = total_intensity_labeled[total_intensity_labeled['Cancer Type'].isin(desired_types)]

# Plot histograms in the same plot with different colors
plt.figure(figsize=(10, 6))
colors = plt.cm.get_cmap('viridis', len(total_intensity_labeled['Cancer Type'].unique()))

for i, (label, color) in enumerate(zip(total_intensity_labeled['Cancer Type'].unique(), colors.colors)):
    subset = total_intensity_labeled[total_intensity_labeled['Cancer Type'] == label]
    subset['total_intensity'].hist(bins=30, alpha=0.5, label=f'Cancer Type {label}', color=color)

plt.xlabel('Total intensity (log10)')
plt.ylabel('Frequency')
plt.title('Distribution of total intensity by Cancer Type')
plt.legend()
plt.show()
```

```python
peptide_level_aggregation = citrullination.groupby(['Run', 'Stripped.Sequence']).agg({'has_target_PTM': 'any', 'num_PTM': 'max'}).reset_index()

protein_level_aggregation = citrullination.groupby(['Run', 'Protein.Group']).agg({'has_target_PTM': 'any', 'num_PTM': 'max'}).reset_index()
```

```python
import matplotlib.pyplot as plt

citrullination_labeled = protein_level_aggregation.merge(metadata, left_on='Run', right_on='PLateNumber')


# Drop rows with missing 'Is_Case' values
citrullination_labeled = citrullination_labeled .dropna(subset=['Is_Case'])

# Calculate the mean of 'has_PTM' for each 'Run' and 'Is_Case'
mean_has_PTM = citrullination_labeled.groupby(['Run', 'Is_Case'])['has_target_PTM'].mean().reset_index()

# Plot histograms in the same plot with different colors
plt.figure(figsize=(10, 6))
for label, color in zip(mean_has_PTM['Is_Case'].unique(), ['blue', 'orange']):
    subset = mean_has_PTM[mean_has_PTM['Is_Case'] == label]
    subset['has_target_PTM'].hist(bins=50, alpha=0.5, label=f'Is_Case {label}', color=color)

plt.xlabel('Proportion of modified peptides')
plt.ylabel('Frequency')
plt.title('Distribution of proportion of modified peptides')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Drop rows with missing 'Is_Case' values
citrullination_labeled = citrullination.dropna(subset=['Is_Case'])

# Calculate the total number of peptides with PTM for each 'Run' and 'Is_Case'
total_peptides_with_PTM = citrullination_labeled.groupby(['Run', 'Is_Case'])['has_target_PTM'].sum().reset_index()

# Plot histograms in the same plot with different colors
plt.figure(figsize=(10, 6))
for label, color in zip(total_peptides_with_PTM['Is_Case'].unique(), ['blue', 'orange']):
    subset = total_peptides_with_PTM[total_peptides_with_PTM['Is_Case'] == label]
    subset['has_target_PTM'].hist(bins=100, alpha=0.5, label=f'Is_Case {label}', color=color)

plt.xlabel('Total number of peptides with PTM')
plt.ylabel('Frequency')
plt.title('Histogram of Total Number of Peptides with PTM by Is_Case')
plt.legend()
plt.show()
```
